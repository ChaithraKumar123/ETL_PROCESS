```python
import pandas as pd
```

```python
data=pd.read_csv('Crashes_Last_Five_Years.csv')
```

```python
data.columns
```

```python
df=data[['ACCIDENT_DATE','ALCOHOLTIME','ACCIDENT_TYPE','ROAD_GEOMETRY','SEVERITY','SPEED_ZONE','TOTAL_PERSONS','INJ_OR_FATAL',
      'DRIVER','OLD_DRIVER','YOUNG_DRIVER','ALCOHOL_RELATED']]
```

```python
df.columns
```

```python
df['Year']=df['ACCIDENT_DATE'].str.split('/').apply(lambda x: ('20'+str(x[2])))
```

```python
df
```

```python
df=df.groupby(['Year','ACCIDENT_TYPE','ROAD_GEOMETRY','SEVERITY','SPEED_ZONE','ALCOHOL_RELATED']).agg({'TOTAL_PERSONS':'sum','INJ_OR_FATAL':'sum','DRIVER':'sum','OLD_DRIVER':'sum','YOUNG_DRIVER':'sum'}).reset_index()
```

```python
df.to_csv('Crashes_Last_Five_Years_etl.csv')
```

```python
data_qz=pd.read_csv('factorsinroadcrashes.csv')
```

```python
data_qz=data_qz[data_qz['Crash_Year'].isin(['2013', '2014', '2015', '2016', '2017', '2018', '2019'])]
```

```python
data_qz=data_qz[data_qz['Crash_Police_Region']=='South Eastern']
```

```python
data_qz.groupby(['Crash_Year','Crash_Severity','Involving_Drink_Driving','Involving_Driver_Speed','Involving_Fatigued_Driver','Involving_Defective_Vehicle']).agg({'Count_Crashes':'sum'}).reset_index()
```

```python
data_qz.to_csv('factorsinroadcrashes_etl.csv')
```
